```python
from rdflib import Graph
from rdflib.namespace import RDF

# =========================
# Chargement du fichier Turtle
# =========================
graphe = Graph()
graphe.parse("etab.ttl", format="turtle")

print(f"Graphe chargé avec {len(graphe)} triplets\n")

# Préfixes SPARQL
PREFIXES = """
PREFIX : <http://example.org/universite#>
PREFIX rdf: <http://www.w3.org/1999/02/22-rdf-syntax-ns#>
"""

# =========================
# Requête 1 : Tous les apprenants
# =========================
rq_apprenants = PREFIXES + """
SELECT ?a
WHERE {
    ?a rdf:type :Apprenant .
}
"""

print("Liste des apprenants :")
for res in graphe.query(rq_apprenants):
    print(" -", res.a)

# =========================
# Requête 2 : Tous les enseignants
# =========================
rq_enseignants = PREFIXES + """
SELECT ?e
WHERE {
    ?e rdf:type :Enseignant .
}
"""

print("\n Liste des enseignants :")
for res in graphe.query(rq_enseignants):
    print(" -", res.e)

# =========================
# Requête 3 : Enseignants et cours assurés
# =========================
rq_cours = PREFIXES + """
SELECT ?e ?c
WHERE {
    ?e :donneCours ?c .
}
"""

print("\n Enseignants et leurs cours :")
for res in graphe.query(rq_cours):
    print(f" - {res.e} donne le cours {res.c}")

# =========================
# Requête 4 : Apprenants et université
# =========================
rq_universite = PREFIXES + """
SELECT ?a ?u
WHERE {
    ?a :etudieA ?u .
}
"""

print("\n Apprenants et leur université :")
for res in graphe.query(rq_universite):
    print(f" - {res.a} étudie à {res.u}")

# =========================
# Requête 5 : Tous les cours
# =========================
rq_liste_cours = PREFIXES + """
SELECT ?c
WHERE {
    ?c rdf:type :Cours .
}
"""

print("\n Liste des cours disponibles :")
for res in graphe.query(rq_liste_cours):
    print(" -", res.c)
```
